```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data.csv')
```

```python
# Defining the features and target
feature = np.array(data['feature'], np.float32)

target = np.array(data['target'], np.float32)
```

```python
# Define the intercept and the slope

intercept = tf.Variable(0.1, np.float32)
slope = tf.Variable(0.1, np.float32)
```

```python
# Define a linear regression model

def linear_regression(intercept, slope, features = feature):
    return intercept + slope*features
```

```python
# Compute the predicted values and the loss

def loss(intercept, slope, targets = target, features= feature):
    predictions = linear_regression(intercept, slope)
    
    return tf.keras.losses.mse(targets, predictions)
```

```python
# Define an optimizer
opt = tf.keras.optimizers.Adam()        
```

```python
# Minimize the loss function and print the loss
for j in range(10000):
    opt.minimize(lambda : loss(intercept, slope),\
    var_list = [intercept, slope])
    print(loss(intercept, slope))
```
